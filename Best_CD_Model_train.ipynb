```python
model = YOLO('yolov8n.pt')

results = model.train(
    data='/content/dataset/trafic_data/data_1.yaml',
    epochs=30,
    imgsz=640,
    batch=16,
    name='custom_yolov8_model'
)
```

```python
from ultralytics import YOLO

# Load the trained model
model = YOLO('/content/runs/detect/custom_yolov8_model2/weights/best.pt')

# Path of your test image
test_image = '/content/64_jpg.rf.e88dfd8566cbc446c44650ea0b13d8d3.jpg'

# Run prediction
results = model.predict(
    source=test_image,
    show=True,
    save=True
)

# Show where result is saved
print("Prediction complete!")
print("Results saved in: /content/runs/detect/predict/")
```
